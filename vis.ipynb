```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.manifold import TSNE
import colorcet as cc
import matplotlib.pyplot as plt
```

```python

with open('./sec_last_for_tsne_pred_87.pk','rb') as f:
    pred = pickle.load(f)
print(np.array(pred)[:,0,:].shape)
x = np.array(pred)[:,0,:]
```

```python
with open('./sec_last_for_tsne_label_87.pk','rb') as f:
    labels = pickle.load(f)
print(np.array(labels))
print(np.array(labels)[::-1].shape)
```

```python

x = StandardScaler().fit_transform(x)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf['target'] = np.array(labels)
```

```python
print(principalDf)
```

```python
sns.lmplot( x="principal component 1", y="principal component 2",
  data=principalDf, 
  fit_reg=False, 
  hue='target', # color by cluster
  palette = sns.color_palette(cc.glasbey, n_colors=50),
  legend=True,
  scatter_kws={"s": 80}) # specify the point size
ax = plt.gca()
ax.set_title("pca of first epoch")
plt.savefig('./pca', dpi=300, bbox_inches='tight')
```

```python
tsne = TSNE(n_components=2)
tsne_component = tsne.fit_transform(x)
```

```python
tsneDf = pd.DataFrame(data = tsne_component
             , columns = ['tsne1', 'tsne2'])
tsneDf['target'] = np.array(labels)
```

```python
sns.lmplot(x="tsne1", y="tsne2",
  data=tsneDf, 
  fit_reg=False, 
  hue='target', # color by cluster
  palette = sns.color_palette(cc.glasbey, n_colors=50),
  legend=True,
  scatter_kws={"s": 80}) # specify the point size
ax = plt.gca()
ax.set_title("tsne of last epoch")
plt.savefig('./tsne_last', dpi=300, bbox_inches='tight')
```
